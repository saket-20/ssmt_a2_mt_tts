```python
import sacrebleu

# Load the reference and hypothesis files
with open("/content/test_actual_telugu.txt", "r") as ref_file:
    references = ref_file.read().splitlines()

with open("/content/test_tel_pred.txt", "r") as pred_file:
    predictions = pred_file.read().splitlines()

# Compute the BLEU score
bleu = sacrebleu.corpus_bleu(predictions, [references])

# Print the BLEU score
print(f"BLEU score: {bleu.score:.2f}")

# Find and print 5 incorrect translations with their line numbers
print("\nIncorrectly Translated Samples (up to 5):")
incorrect_count = 0

for i, (ref, pred) in enumerate(zip(references, predictions)):
    if ref != pred:
        print(f"\nSample {i+1}: (Line {i+1})")  # Line numbers are 1-based
        print(f"Reference: {ref}")
        print(f"Prediction: {pred}")
        incorrect_count += 1

    # Stop after printing 5 incorrect examples
    if incorrect_count == 5:
        break

if incorrect_count == 0:
    print("No incorrect translations found.")
```

```python
from gtts import gTTS
import os

# Specify the path to the text file
text_file = 'test_tel_pred.txt'

# Open the text file and read it line by line
with open(text_file, 'r', encoding='utf-8') as file:
    lines = file.readlines()

# Iterate over the first 5 lines, convert to audio, and save the file
for i, line in enumerate(lines[:15]):  # Limit to the first 5 lines
    # Remove any leading/trailing whitespace
    line = line.strip()

    if line:
        # Generate speech for the line (language code 'te' for Telugu)
        tts = gTTS(text=line, lang='te')

        # Save the audio file (e.g., 'audio_0.mp3', 'audio_1.mp3', etc.)
        output_file = f'audio_{i}.mp3'
        tts.save(output_file)
        print(f"Saved {output_file}")
```

```python
from transformers import pipeline
import numpy as np
import soundfile as sf

synthesizer = pipeline("text-to-speech", model="facebook/mms-tts-tel")
text = "."
result = synthesizer(text)

# Debug: print output keys and sampling rate
print("Output keys:", result.keys())
print("Sampling rate:", result["sampling_rate"])

# Get the audio data (a list) and convert it to a NumPy array
audio_data = result["audio"]
audio_array = np.asarray(audio_data, dtype=np.float32)

# Normalize the audio to the range of int16
if np.max(np.abs(audio_array)) != 0:
    audio_int16 = np.int16(audio_array / np.max(np.abs(audio_array)) * 32767)
else:
    audio_int16 = np.int16(audio_array)

# Save the normalized audio as a WAV file using the sampling rate
sf.write("output_audio.wav",audio_int16, int(result["sampling_rate"]))
```
